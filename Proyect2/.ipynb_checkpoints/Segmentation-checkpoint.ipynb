### Extracting binary mask with thresholding
The first step is to extract all bubbles from the background.
For that, we will first convert the image to grayscale with $ cv2.Color() $ function, and use $ cv2.threshold() $ to convert it to binary image.

- For this input the threshold is set to **60**, so all pixels below 60 are set to 0 and the others are set to 255, 
- Since some of the bubbles are slightly overlapped on the binary image, we use the $ cv2.erode() $ function to separate them.




```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
```

```python
im = cv2.imread('img/bubble.png')
im2 = im[:,:,::-1]
plt.imshow(im2)
plt.axis('off')
```

```python
gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
_, msk = cv2.threshold(gray, 60, 255, cv2.THRESH_BINARY)
msk = cv2.erode(msk, np.ones((3,3), np.uint8))
plt.imshow(msk, cmap = 'gray')
```

### Extracting objects borders using Contours

The next step is to find objects in the binary image, we use $ cv2.findContours() $ function on the binary image to detect the border,
countour is define continuos curve in the boundary of an image.


```python
contours, _ = cv2.findContours(msk, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
#The flag CHAIN_APPROX_SIMPLE only return extern contours.

contours_img_before_filtering = msk.copy()
contours_img_before_filtering = cv2.cvtColor(contours_img_before_filtering, cv2.COLOR_GRAY2BGR)
cv2.drawContours(contours_img_before_filtering, contours, -1, (0, 255, 0), 1)
plt.imshow(contours_img_before_filtering)
plt.axis('off')

```

### Filtering Contours and Extracting Mean Colors
- Remove countours that do not represent a bubbles, we will iterate for all resulting contours and select only those with area greater that 3000, so discard any other object.
- To find the mean color of each bubble, we first create a mask for each bubble and set the drawing its contours in white and other in black.
- The $ cv2.mean() $ function to calculate the bubble's mean Blue, Green, Red and stored in a DataFrame.

```python
filtered_contours = []
df_mean_color = pd.DataFrame()

cnt = 0
for idx, contour in enumerate(contours) :
    area = int(cv2.contourArea(contour))
    if area < 100 and area > 20 :
        filtered_contours.append(contour)
        masked = np.zeros_like(im[:,:,0])
        cv2.drawContours(masked, [contour], 0, 255, -1)
        B_mean, G_mean, R_mean, _ = cv2.mean(im, mask=masked)
        df = pd.DataFrame({'B_mean': B_mean, 'G_mean': G_mean, 'R_mean': R_mean}, index=[idx])
        df_mean_color = pd.concat([df_mean_color, df])

df_mean_color.head()
```

```python
#Displat filtered contours
contours_img_after_filtering = msk.copy()
contours_img_after_filtering = cv2.cvtColor(contours_img_after_filtering, cv2.COLOR_GRAY2BGR)
cv2.drawContours(contours_img_after_filtering, tuple(filtered_contours), -1, (0, 255, 0), 1)
plt.imshow(contours_img_after_filtering)
plt.axis('off')
```

### Clustering similar colors with K-mean algorithm 
We will use mean color for each bubble to input data for the **KMeans** algorithm from $ sklearn \ library$.
In this case, since there are 6 bubbles colors, we'll set the value to 6.


```python
km = KMeans( n_clusters=6)
df_mean_color['label'] = km.fit_predict(df_mean_color)
```

```python
def draw_segmented_objects(image, countour, label_cnt_idx, bubbles_count):
    mask = np.zeros_like(image[:, :, 0])
    cv2.drawContours(mask, [contours[i] for i in label_cnt_idx], -1, (255), -1)
    masked_image = cv2.bitwise_and(image, image, mask=mask)
    masked_image = cv2.putText(masked_image, f'{bubbles_count} bubbles', (200, 1200), cv2.FONT_HERSHEY_SIMPLEX, 
                               fontScale = 3, color = (255, 255, 255), thickness = 10, lineType = cv2.LINE_AA)
    return masked_image
```

```python
img = im.copy()

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB) )
plt.axis('off')
plt.show()
for label, df_grouped in df_mean_color.groupby('label') :
    bubbles_amount = len(df_grouped)
    masked_image = draw_segmented_objects(im, contour, df_grouped.index, bubbles_amount)
    #img = cv2.hconcat([img, masked_image])
    plt.imshow(cv2.cvtColor(masked_image, cv2.COLOR_BGR2RGB) )
    plt.axis('off')
    plt.show()


cv2.imwrite('color_segmentation.png', img)
```
